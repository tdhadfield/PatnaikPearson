# Experiment One

```python
N = 2000
d = 1000
initial_alpha_X0 = 2.0
initial_alpha_X1 = 3.0

xx = pp.interpolation_experiment(N, d, initial_alpha_X0, initial_alpha_X1)
```

```python
results_dict = xx

t_vals = results_dict["t_vals"]
initial_alpha_X0_vals = results_dict["initial_alpha_X0_vals"]
initial_alpha_X1_vals = results_dict["initial_alpha_X1_vals"]
actual_alpha_X0_vals = results_dict["actual_alpha_X0_vals"]
actual_alpha_X1_vals = results_dict["actual_alpha_X1_vals"]
actual_alpha_Xt_vals = results_dict["actual_alpha_Xt_vals"]
nu_over_d_X0_vals = results_dict["nu_over_d_X0_vals"]
nu_over_d_X1_vals = results_dict["nu_over_d_X1_vals"]
nu_over_d_Xt_vals = results_dict["nu_over_d_Xt_vals"]
estimate_alpha_Xt_vals = results_dict["estimate_alpha_Xt_vals"]
estimate_nu_over_d_Xt_vals = results_dict["estimate_nu_over_d_Xt_vals"]
```

```python
this_title = "alpha(Xt) for Xt = (1-t)*X0 + t*X1, as t varies between 0 and 1\n"
actual_alpha_X0 = actual_alpha_X0_vals[0]
actual_alpha_X1 = actual_alpha_X1_vals[0]
str_alpha_X0 = str(0.001 * int(1000 * actual_alpha_X0))
str_alpha_X1 = str(0.001 * int(1000 * actual_alpha_X1))
this_title += "alpha(X0) = " + str_alpha_X0 + ", alpha(X1) = " + str_alpha_X1
this_title += ", N = " + str(N) + ", d = " + str(d)
plt.plot(t_vals, actual_alpha_Xt_vals, label="alpha(Xt)")
plt.plot(t_vals, actual_alpha_X0_vals, label = "alpha(X0)")
plt.plot(t_vals, actual_alpha_X1_vals, label = "alpha(X1)")
plt.xlabel("t")
plt.ylabel("alpha")
plt.legend()
plt.title(this_title)
#plt.savefig('interpolation_alpha_Xt_alpha_X0_0806_alpha_X1_2275_as_t_varies.pdf', dpi=300, bbox_inches='tight')
plt.show()


this_title = "nu/d (Xt) for Xt = (1-t)*X0 + t*X1, as t varies between 0 and 1\n"
nu_over_d_X0 = nu_over_d_X0_vals[0]
nu_over_d_X1 = nu_over_d_X1_vals[0]
str_nu_over_d_X0 = str(0.001 * int(1000 * nu_over_d_X0))
str_nu_over_d_X1 = str(0.001 * int(1000 * nu_over_d_X1))
this_title += "nu/d (X0) = " + str_nu_over_d_X0 + ", nu/d (X1) = " + str_nu_over_d_X1
this_title += ", N = " + str(N) + ", d = " + str(d)
plt.plot(t_vals, nu_over_d_Xt_vals, label="nu/d (Xt)")
plt.plot(t_vals, nu_over_d_X0_vals, label = "nu/d (X0)")
plt.plot(t_vals, nu_over_d_X1_vals, label = "nu/d (X1)")
plt.xlabel("t")
plt.ylabel("nu/d")
plt.legend()
plt.title(this_title)
#plt.savefig('interpolation_nu_over_d_Xt_nu_over_d_X0_0016_nu_over_d_X1_0651_as_t_varies.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
